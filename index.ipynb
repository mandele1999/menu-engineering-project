## Menu Analysis & Classification

Let's start by getting item-level summaries

```python
# Convert raw transaction data into item-level summaries
menu_summary = (
    df.groupby(['item_name', 'category', 'unit_price', 'cost_price'])
    .agg({
        'quantity_sold': 'sum',
        'revenue': 'sum',
        'gross_profit': 'sum'
    })
    .reset_index()
)
```

```python
print("Menu Summary Shape: ", menu_summary.shape)
menu_summary
```

```python
# Compute derived metrics
menu_summary["profit_per_item"] = menu_summary["unit_price"] - menu_summary["cost_price"]
menu_summary["total_sales_volume"] = menu_summary["quantity_sold"]
menu_summary["revenue_contribution_%"] = (menu_summary["revenue"] / menu_summary["revenue"].sum()) * 100
menu_summary["profit_margin_%"] = (menu_summary["profit_per_item"] / menu_summary["unit_price"]) * 100

# Select relevant columns for inspection
menu_summary = menu_summary[[
    "item_name", "category", "unit_price", "cost_price", "profit_per_item",
    "total_sales_volume", "revenue", "revenue_contribution_%",
    "gross_profit", "profit_margin_%"
]]

menu_summary.sort_values(by="total_sales_volume", ascending=False).head()
```

**Thresholds for Classification**

This is the decision rule to put items into 4 quadrants

We'll be using the median of `total_sales_volume` and `profit_per_item` as cutoff points.

```python
# Plot distribution of profit per item 
plt.figure(figsize=(12, 6))
sns.histplot(menu_summary["profit_per_item"], bins=5, kde=True)
plt.title("Distribution of Profit per Item")
plt.xlabel("Profit per Item")
plt.ylabel("Frequency")
plt.axvline(menu_summary["profit_per_item"].mean(), color='red', linestyle='--', label='Mean Profit')
plt.axvline(menu_summary["profit_per_item"].median(), color='green', linestyle='--', label='Median Profit')
plt.legend()
plt.show()
```

```python
# Plot distribution of sales volume
plt.figure(figsize=(12, 6))
sns.histplot(menu_summary["total_sales_volume"], bins=5, kde=True)
plt.title("Distribution of Total Sales Volume")
plt.xlabel("Total Sales Volume")
plt.ylabel("Frequency")
plt.axvline(menu_summary["total_sales_volume"].mean(), color='red', linestyle='--', label='Mean Sales Volume')
plt.axvline(menu_summary["total_sales_volume"].median(), color='green', linestyle='--', label='Median Sales Volume')
plt.legend()
plt.show()
```

```python
# Define thresholds for classification
sales_threshold = menu_summary["total_sales_volume"].median()
profit_threshold = menu_summary["profit_per_item"].median()

# Classification logic
def classify(row):
    if row["total_sales_volume"] >= sales_threshold and row["profit_per_item"] >= profit_threshold:
        return "Star"
    elif row["total_sales_volume"] >= sales_threshold and row["profit_per_item"] < profit_threshold:
        return "Plowhorse"
    elif row["total_sales_volume"] < sales_threshold and row["profit_per_item"] >= profit_threshold:
        return "Puzzle"
    else:
        return "Dog"

# Apply classification
menu_summary["classification"] = menu_summary.apply(classify, axis=1)

# Show classification result
menu_summary[["item_name", "total_sales_volume", "profit_per_item", "classification"]].sort_values(by="classification")
```

```python
# Create the 2x2 plot
plt.figure(figsize=(14, 7))
plot = sns.scatterplot(
    data=menu_summary,
    x="total_sales_volume",
    y="profit_per_item",
    hue="classification",
    style="classification",
    s=100,
    palette="Set2"
)

# Plot threshold lines
plt.axvline(sales_threshold, color='gray', linestyle='--', label="Sales Threshold")
plt.axhline(profit_threshold, color='gray', linestyle='--', label="Profit Threshold")

# Annotate points
for i in range(menu_summary.shape[0]):
    item = menu_summary.iloc[i]
    plt.text(
        item["total_sales_volume"] + 1,
        item["profit_per_item"] + 2,
        item["item_name"],
        fontsize=9,
        alpha=0.8
    )

# Set labels and title
plt.title("Menu Engineering Matrix", fontsize=14, weight='bold')
plt.xlabel("Total Sales Volume")
plt.ylabel("Profit per Item")
plt.legend(title="Classification")
plt.tight_layout()
plt.show()
```

```python
# Define suggested actions based on classification
def suggest_action(classification):
    return {
        "Star": "Promote heavily and maintain quality",
        "Plowhorse": "Reduce cost or adjust pricing",
        "Puzzle": "Rebrand or pair with offers",
        "Dog": "Consider removing or replacing"
    }.get(classification, "")

menu_summary["suggested_action"] = menu_summary["classification"].apply(suggest_action)
```

## Time-Based Trend Analysis

We'll create a new column to flag weekdays vs weekends, then group and analyze accordingly.

```python
# Plot average sales for month june
plt.figure(figsize=(12, 6))
monthly_sales = df.groupby(df['date'].dt.to_period('D')).agg({'revenue': 'sum'}).reset_index()
monthly_sales['date'] = monthly_sales['date'].dt.to_timestamp()
plt.plot(monthly_sales['date'], monthly_sales['revenue'], marker='o', linestyle='-', color='b', label='Daily Revenue')
# Add trend line
z = np.polyfit(monthly_sales['date'].map(pd.Timestamp.toordinal), monthly_sales['revenue'], 1)
p = np.poly1d(z)
plt.plot(
    monthly_sales['date'],
    p(monthly_sales['date'].map(pd.Timestamp.toordinal)),
    color='red',
    linestyle='--',
    label='Trend Line'
)
plt.title('Average Sales for June 2023')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
df["date"] = pd.to_datetime(df["date"])
df["day_name"] = df["date"].dt.day_name()
df["day_of_week"] = df["date"].dt.dayofweek  # Monday = 0, Sunday = 6
df["day_type"] = df["day_of_week"].apply(lambda x: "Weekend" if x >= 5 else "Weekday")
```

```python
trend_summary = (
    df.groupby(["item_name","day_type"])
    .agg({
        "quantity_sold": "sum",
        "revenue": "sum",
        "gross_profit": "sum"
    })
    .reset_index()
)
```

```python
pivot_trends = trend_summary.pivot(index="item_name", columns="day_type", values="quantity_sold").fillna(0)
```

```python
# Count how many weekdays and weekends in the data
weekend_days = df[df["day_type"] == "Weekend"]["date"].nunique()
weekday_days = df[df["day_type"] == "Weekday"]["date"].nunique()

# Normalize quantity sold per day
pivot_trends["Weekend per Day"] = pivot_trends["Weekend"] / weekend_days
pivot_trends["Weekday per Day"] = pivot_trends["Weekday"] / weekday_days
pivot_trends["weekend_lift_%"] = ((pivot_trends["Weekend per Day"] - pivot_trends["Weekday per Day"]) / pivot_trends["Weekday per Day"]) * 100
pivot_trends = pivot_trends.sort_values(by="weekend_lift_%", ascending=False)
```

### Scenario Simulation

Simulation 1:

```python
# Scenario Simulation
scenario = menu_summary.copy()
scenario.loc[scenario["classification"] == "Dog", ["total_sales_volume", "revenue", "gross_profit"]] = 0

boost_idx = scenario["classification"] == "Puzzle"
scenario.loc[boost_idx, "total_sales_volume"] *= 1.20
scenario.loc[boost_idx, "revenue"] = scenario.loc[boost_idx, "unit_price"] * scenario.loc[boost_idx, "total_sales_volume"]
scenario.loc[boost_idx, "gross_profit"] = (scenario.loc[boost_idx, "unit_price"] - scenario.loc[boost_idx, "cost_price"]) * scenario.loc[boost_idx, "total_sales_volume"]

# Compare results
baseline_profit = menu_summary["gross_profit"].sum()
scenario_profit = scenario["gross_profit"].sum()
profit_lift_pct = ((scenario_profit - baseline_profit) / baseline_profit) * 100

print("Baseline Gross Profit: ", baseline_profit) 
print("Simulated Gross Profit: ", scenario_profit)
print("Change: ", profit_lift_pct)
```

Instead of increasing profit, the scenario reduced it by 9.7%.
- Boosting Puzzle by 20% wasn't enough the offset the loss that came with removing all Dog items, including their modest contributions

- It's clear some Dogs, while low-performing, still contributed net positive profits.

Simulation 2:

```python
# Simulating an improved scenario:
# - Boost Puzzle and Star items by 20%
# - Increase Plowhorse price by 10%
# - Keep Dogs but reduce their sales by 50% instead of removing

# Start fresh from the original menu summary
improved_scenario = menu_summary.copy()

# Adjust Dog sales down by 50%
dog_idx = improved_scenario["classification"] == "Dog"
improved_scenario.loc[dog_idx, "total_sales_volume"] *= 0.5
improved_scenario.loc[dog_idx, "revenue"] = improved_scenario.loc[dog_idx, "unit_price"] * improved_scenario.loc[dog_idx, "total_sales_volume"]
improved_scenario.loc[dog_idx, "gross_profit"] = (improved_scenario.loc[dog_idx, "unit_price"] - improved_scenario.loc[dog_idx, "cost_price"]) * improved_scenario.loc[dog_idx, "total_sales_volume"]

# Boost Puzzle and Star sales by 20%
boost_idx = improved_scenario["classification"].isin(["Puzzle", "Star"])
improved_scenario.loc[boost_idx, "total_sales_volume"] *= 1.20
improved_scenario.loc[boost_idx, "revenue"] = improved_scenario.loc[boost_idx, "unit_price"] * improved_scenario.loc[boost_idx, "total_sales_volume"]
improved_scenario.loc[boost_idx, "gross_profit"] = (improved_scenario.loc[boost_idx, "unit_price"] - improved_scenario.loc[boost_idx, "cost_price"]) * improved_scenario.loc[boost_idx, "total_sales_volume"]

# Increase Plowhorse price by 10% (and recompute profit)
plow_idx = improved_scenario["classification"] == "Plowhorse"
improved_scenario.loc[plow_idx, "unit_price"] *= 1.10
improved_scenario.loc[plow_idx, "revenue"] = improved_scenario.loc[plow_idx, "unit_price"] * improved_scenario.loc[plow_idx, "total_sales_volume"]
improved_scenario.loc[plow_idx, "gross_profit"] = (improved_scenario.loc[plow_idx, "unit_price"] - improved_scenario.loc[plow_idx, "cost_price"]) * improved_scenario.loc[plow_idx, "total_sales_volume"]

# Recalculate profit under improved scenario
improved_profit = improved_scenario["gross_profit"].sum()
improved_lift_pct = ((improved_profit - baseline_profit) / baseline_profit) * 100

baseline_profit, improved_profit, improved_lift_pct
```

- Boosting Stars + Puzzles by 20% → amplifies already profitable or high-potential items
- Increasing Plowhorse prices by 10% → leverages high-volume items with thin margins
- Halving Dog sales instead of deleting → you keep some cash flow from them without dragging down profitability

```python
# Marketing cost for boosted items
boosted_items = improved_scenario[improved_scenario["classification"].isin(["Star", "Puzzle"])]
num_boosted_items = len(boosted_items)
marketing_cost_per_item = 3000
total_marketing_cost = num_boosted_items * marketing_cost_per_item

# Net profit
net_profit_after_marketing = improved_profit - total_marketing_cost
net_profit_lift_pct = ((net_profit_after_marketing - baseline_profit) / baseline_profit) * 100

num_boosted_items, total_marketing_cost, net_profit_after_marketing, net_profit_lift_pct
```
